## Mengambil data dari API cuaca.bmkg.go.id

```python
import requests
import pandas as pd
```

```python
# Variabel API yang dibutuhkan

api_url="https://cuaca.bmkg.go.id/api/df/v1/forecast/adm"
params = {"adm1":"33",
          "adm2":"33.02"}
```

```python
response = requests.get(api_url, params=params)
api_data = response.json()
```

```python
api_data # cek struktur data
```

```python
kec_data_list = api_data.get("data")
```

```python
kec_data_list # cek struktur data
```

```python
kec_data_list[0] # cek salah satu kecamatan
```

```python
# mengambil informasi lokasi dan forecast dari satu titik kecamatan
kec_data = kec_data_list[0]
lokasi = kec_data.get("lokasi")
data_cuaca = kec_data.get("cuaca")
```

```python
# memisahkan data harian menjadi per waktu
data_container = []
for hari in data_cuaca:
  for waktu in hari:
    data_container.append(waktu)

data_container
```

```python
# membuat dataframe dari data container

df = pd.json_normalize(data_container)
df
```

```python
# menambahkan data lokasi
for key, value in lokasi.items():
  df[key] = value

df
```

```python
# menyimpan dataframe kedalam csv dengan format nama file berdasarkan lokasi

prov = lokasi.get('provinsi')
kab = lokasi.get('kotkab')
kec = lokasi.get('kecamatan')

filename = f"{prov}_{kab}_{kec}.csv"
df.to_csv(filename, index=False)
```

## Membuat DataFrame berisi seluruh kecamatan

```python
def get_data_from_api(api_url, headers=None, params=None):
    """
    Fetch data from an API.

    Args:
        api_url (str): The endpoint of the API.
        headers (dict): Optional headers to include in the request (e.g., for authentication).
        params (dict): Optional query parameters for the request.

    Returns:
        dict: Parsed JSON data from the API response.
        None: If the request fails or an error occurs.
    """
    try:
        # Make a GET request to the API
        response = requests.get(api_url, headers=headers, params=params)
        # Raise an error for bad responses (4xx and 5xx)
        response.raise_for_status()

        # Parse and return the JSON response
        return response.json()
    except requests.exceptions.RequestException as e:
        # Handle request exceptions
        print(f"Error fetching data from API: {e}")
        return None

def make_kecamatan_dataframe(kec_data):
  lokasi = kec_data.get("lokasi")
  data_cuaca = kec_data.get("cuaca")

  data_container = []
  for hari in data_cuaca:
    for waktu in hari:
      data_container.append(waktu)

  df = pd.json_normalize(data_container)

  for key, value in lokasi.items():
    df[key] = value
  return df
```

```python
api_url="https://cuaca.bmkg.go.id/api/df/v1/forecast/public/adm"
params = {"adm1":"33",
          "adm2":"33.02"}
api_data = get_data_from_api(api_url, params=params)
kec_data_list = api_data.get("data")

df_container = []
for kec_data in kec_data_list:
  df = make_kecamatan_dataframe(kec_data)
  df_container.append(df)

df_all = pd.concat(df_container, ignore_index=True)

df_all.to_csv("forecast_banyumas.csv",sep=",",index=False)

```
