```python
import folium
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('/content/forecast_banyumas.csv')
df
```

```python
#filter data sesuai tanggal
tanggal = '2024-11-20T03:00:00Z'
df_filter = df[df['datetime'] == tanggal]
df_filter
```

```python
map_center = [-7.421693731245026, 109.16612740697146]
my_map = folium.Map(location=map_center, zoom_start=13)
my_map
```

```python
map_center = [-7.421693731245026, 109.16612740697146]
my_map = folium.Map(location=map_center, zoom_start=13)
for index, row in df_filter.iterrows():
    latitude = row['lat']
    longitude = row['lon']
    icon_url = row['image']
    desc = row['weather_desc']

    weather_icon = folium.CustomIcon(icon_image=icon_url, icon_size=(50, 50))
    folium.Marker(
        location=[latitude, longitude],
        icon=weather_icon,
        popup = desc
    ).add_to(my_map)

my_map
```

```python
map_center = [-7.421693731245026, 109.16612740697146]
my_map = folium.Map(location=map_center, zoom_start=13)
for index, row in df_filter.iterrows():
    latitude = row['lat']
    longitude = row['lon']
    icon_url = row['image']
    desc = row['weather_desc']

    temp = row['t']
    tcc = row['tcc']
    tp = row['tp']
    hu = row['hu']
    kec = row['kecamatan']

    pop_content = f'''
    <div style="font-family: Arial; font-size: 14px; color: darkblue;">
    <h4>{kec}</h4>
    <p><img src={icon_url} alt="Weather Icon" width="30"></p>
    <p>{desc}</p>
    <p><strong>Temperature:</strong> {temp} °C</p>
    <p><strong>Humidity:</strong> {hu} %</p>
    <p><strong>Cloud:</strong> {tcc} %</p>
    </div>
    '''
    custom_popup = folium.Popup(pop_content, max_width=300)

    weather_icon = folium.CustomIcon(icon_image=icon_url, icon_size=(50, 50))
    folium.Marker(
        location=[latitude, longitude],
        icon=weather_icon,
        popup = custom_popup
    ).add_to(my_map)

my_map
```
